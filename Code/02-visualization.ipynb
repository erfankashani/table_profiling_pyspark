# Visualizing the results

In this notebook we assume the results are generated in json format and we aim to visualize the results using jija2

```python
# Libraries

from jinja2 import Template, Environment, FileSystemLoader
import json
```

```python
# variables
result_path = '/Users/erfankashani/code/src/github.com/erfan_github/Spark_profiling/Result/COVID_19_vs_Vaccine_in_Turkey.csv_profiled_table_2021_09_15-06/59/35_PM.json'
result_json = '../Result/COVID_19_vs_Vaccine_in_Turkey_profiled_table_test.json'
```

```python
# read Json file

with open(result_json) as f:
  data = json.load(f)
```

```python
# simple template
name = input("Enter your name: ")

tm = Template("Hello {{ name }}")
msg = tm.render(name=name)

print(msg)
```

```python
# display the json data

tm_data = Template('{{ json_data }}')
tm_render = tm_data.render(json_data = data)
# print(tm_render)
```

```python
# show the dict using html
file_loader = FileSystemLoader('templates')
env = Environment(loader=file_loader)
env.trim_blocks = True
env.lstrip_blocks = True
env.rstrip_blocks = True

template = env.get_template('temp_test.html')

output = template.render(json_data = data)
print(output)
```

```python
# show the dict in a table html
file_loader = FileSystemLoader('templates')
env = Environment(loader=file_loader)
env.trim_blocks = True
env.lstrip_blocks = True
env.rstrip_blocks = True

template = env.get_template('show_table_from_dict.html')

output = template.render(json_data = data)
print(output)
```

```python
# show the column analysis in html
file_loader = FileSystemLoader('templates')
env = Environment(loader=file_loader)
env.trim_blocks = True
env.lstrip_blocks = True
env.rstrip_blocks = True

template = env.get_template('show_colmn_analysis.html')

output = template.render(json_data = data)
print(output)
```
